```python
import pandas as pd
```

```python
data=pd.read_pickle('sister_cities.pkl', compression='infer')
data
```

# New section

```python
df=pd.DataFrame(data)
df
```

```python
df=df.dropna(axis=1)
df
```

```python
import networkx as nx
G = nx.Graph()
newlist=[]
for col in df.columns[:30]:
    for sis in df.loc['sister_city', col]:
        G.add_edge(col, sis, weight=1)
        

        
```

```python
import scipy
nx.draw(G)
```

```python
from pyvis.network import Network
net=Network(notebook=True)
net.from_nx(G)
net.save_graph("graph.html")
```
